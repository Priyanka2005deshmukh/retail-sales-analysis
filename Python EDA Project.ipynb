```python
# percentage of missing values per column
missing_percentage = (df.isna().sum() / len(df)) * 100

# round to 2 decimals
missing_percentage = missing_percentage.round(2)

print(missing_percentage)
```

```python
#  How many duplicate order IDs are present?
duplicate_order_ids = df['order_id'].duplicated().sum()
print("Duplicate order IDs:", duplicate_order_ids)

# 5 How many invalid or negative quantities exist?
invalid_quantities= (df['quantity'] <= 0).sum()
invalid_quantities
```

```python
df[df['order_date'].isna()]
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
missing_or_invalid_dates = df['order_date'].isna().sum()
print("Records with missing or incorrect dates:", missing_or_invalid_dates)
```

```python
#  List all inconsistent city name variations.
city_variations = df['city'].dropna().unique()
print(city_variations)
```

```python
# 8 Which rows contain extreme price outliers?
Q1 = df['unit_price'].quantile(0.25)
Q3 = df['unit_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['unit_price'] < lower_bound) | (df['unit_price'] > upper_bound)]
outliers
```

```python
# 9 How many rows have missing revenue values?
missing_revenue_count = df['revenue'].isna().sum()
print("Missing revenue rows:", missing_revenue_count)
```

```python
# 10 After cleaning, how many valid transactions remain
clean_df = df.dropna(subset=['order_date', 'revenue'])
clean_df = clean_df[clean_df['quantity'] > 0]
clean_df = clean_df.drop_duplicates(subset='order_id')

print("Valid transactions after cleaning:", len(clean_df))
```

```python
# What is the total revenue generated?
total_revenue = df['revenue'].sum().round(2)
total_revenue
```

```python
# 2 What is the average order value?
avg_order_value = df.groupby('order_id')['revenue'].sum().mean().round(2)
avg_order_value
```

```python
# 3 Which city generates the highest revenue?
city_revenue = df.groupby('city')['revenue'].sum()
top_city = city_revenue.idxmax(), city_revenue.max()
top_city
```

```python
# 4 Which region performs best overall?
region_revenue = df.groupby('region')['revenue'].sum()
best_region = region_revenue.idxmax(), region_revenue.max()
best_region
```

```python
# 5 What are the top 10 selling products by revenue?
top_products = (
    df.groupby('product')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_products
```

```python
# 6 Which product category contributes most to revenue?
category_revenue = df.groupby('category')['revenue'].sum()
top_category = category_revenue.idxmax(), category_revenue.max()
top_category
```

```python
# 7 What is the least performing category?
least_category = category_revenue.idxmin(), category_revenue.min()
least_category
```

```python
# 8 What is the median quantity per order?
median_quantity = df.groupby('order_id')['quantity'].sum().median()
median_quantity
```

```python
# 9 Which payment method is used most frequently?
top_payment_method = df['payment_method'].value_counts().idxmax()
top_payment_method
```

```python
# 10 What percentage of sales comes from online vs in-store?
sales_channel_pct = (
    df.groupby('sales_channel')['revenue']
    .sum()
    .pipe(lambda x: x * 100 / x.sum())
    .round(2)
)
sales_channel_pct
```

```python
# What is the monthly revenue trend?
monthly_revenue = (
    df.groupby(df['order_date'].dt.to_period('M'))['revenue']
    .sum()
    .sort_index()
)
monthly_revenue
```

```python
# 2 Which month has the highest sales?
highest_month = monthly_revenue.idxmax(), monthly_revenue.max()
highest_month
```

```python
# 3 Which month has the lowest sales?
lowest_month = monthly_revenue.idxmin(), monthly_revenue.min()
lowest_month
```

```python
# 4 Is there any seasonality pattern in sales?
df['month'] = df['order_date'].dt.month

seasonality = (
    df.groupby('month')['revenue']
    .mean()
    .sort_index()
)
seasonality
```

```python
# 5 How does average order value change month-to-month?
monthly_aov = (
    df.groupby(df['order_date'].dt.to_period('M'))
      .apply(lambda x: x.groupby('order_id')['revenue'].sum().mean())
)
monthly_aov
```

```python
# 6 Are weekend sales higher than weekday sales?
df['day_type'] = np.where(
    df['order_date'].dt.dayofweek >= 5,
    'Weekend',
    'Weekday'
)

day_type_sales = df.groupby('day_type')['revenue'].sum()
day_type_sales
```

```python
# 7 Which quarter contributes most revenue?
df['quarter'] = df['order_date'].dt.to_period('Q')

quarter_revenue = (
    df.groupby('quarter')['revenue']
    .sum()
    .sort_values(ascending=False)
)
quarter_revenue
```

```python
# Do returning customers spend more than new customers?
avg_spend = (
    df.groupby(['customer_type', 'order_id'])['revenue']
      .sum().round(2)
      .groupby('customer_type')
      .mean().round(2)
)

avg_spend
```

```python
# 2 Which customer type generates more revenue share?
revenue_share = (
    df.groupby('customer_type')['revenue']
      .sum()
      .pipe(lambda x: x * 100 / x.sum())
      .round(2)
)

revenue_share
```

```python
# 3 What is the average quantity per customer type?
revenue_share = (
    df.groupby('customer_type')['revenue']
      .sum()
      .pipe(lambda x: x * 100 / x.sum())
      .round(2)
)

revenue_share
```

```python
# 4 Which city has the highest returning-customer ratio?
city_customer = (
    df.groupby(['city', 'customer_type'])['order_id']
      .nunique()
      .unstack(fill_value=0)
)

city_customer['returning_ratio'] = (
    city_customer['Returning'] /
    city_customer.sum(axis=1)
)

city_customer.sort_values(
    by='returning_ratio',
    ascending=False
).head(1)
```

```python
# 5 Are discounts more common for new customers?
discount_usage = (
    df.assign(discount_used=df['discount_percent'] > 0)
      .groupby('customer_type')['discount_used']
      .mean()
      .mul(100)
      .round(2)
)

discount_usage
```

```python
# What is the average discount percentage overall?
avg_discount = df['discount_percent'].mean()
avg_discount
```

```python
# 2 Which category gives the highest discounts?
category_discount = (
    df.groupby('category')['discount_percent']
      .mean()
      .sort_values(ascending=False)
)

category_discount
```

```python
# 3 Do higher discounts increase revenue?
df['discount_level'] = pd.cut(
    df['discount_percent'],
    bins=[-1, 0, 10, 20],
    labels=['No Discount', 'Low', 'Medium']
)

df.groupby('discount_level', observed=True)['revenue'].mean()
```

```python
# 4 Find correlation between discount, price, and quantity.
correlation_matrix = df[
    ['discount_percent', 'unit_price', 'quantity', 'revenue']
].corr()

correlation_matrix
```

```python
# 5 Identify loss-making or suspicious transactions
loss_orders = df[
    df['final_unit_price'] < (df['unit_price'] * 0.5)
]

loss_orders[['order_id', 'product', 'category', 'discount_percent', 'revenue']]
# “No loss-making transactions were identified, indicating disciplined discounting and strong pricing governance.”
```

```python
#  Which top 5 cities contribute most revenue?
top_5_cities = (
    df.groupby('city')['revenue']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

top_5_cities
```

```python
# 2 Which region has lowest performance?
lowest_region = (
    df.groupby('region')['revenue']
      .sum()
      .sort_values()
      .head(1)
)

lowest_region
```

```python
# 3 Category preference differences across regions.
region_category_pref = (
    df
    .groupby(['region', 'category'])['revenue']
    .sum()
    .reset_index()
    .sort_values(['region', 'revenue'], ascending=[True, False])
)

```

```python
# 4 Which city buys most electronics?
top_electronics_city = (
    df[df['category'] == 'Electronics']
    .groupby('city')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(1)
)

top_electronics_city
```

```python
# 5 Which city prefers fashion products
top_fashion_city = (
    df[df['category'] == 'Fashion']
    .groupby('city')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(1)
)

top_fashion_city
```

```python
# Detect price outliers using IQR or Z-score.
Q1 = df['unit_price'].quantile(0.25)
Q3 = df['unit_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

price_outliers = df[
    (df['unit_price'] < lower_bound) | 
    (df['unit_price'] > upper_bound)
]

price_outliers[['order_id', 'product', 'unit_price']]
# 2 Detect abnormally high revenue orders.
# 3 Are outliers data errors or real premium sales?
# 4 Impact of removing outliers on average revenue.
```

```python
# 2 Detect abnormally high revenue orders.
Q1_rev = df['revenue'].quantile(0.25)
Q3_rev = df['revenue'].quantile(0.75)
IQR_rev = Q3_rev - Q1_rev

upper_rev_limit = Q3_rev + 1.5 * IQR_rev

high_revenue_orders = df[df['revenue'] > upper_rev_limit]

high_revenue_orders[['order_id', 'product', 'quantity', 'revenue']]
```

```python
# 3 Are outliers data errors or real premium sales?
price_outliers.groupby('category')['unit_price'].describe()
```

```python
# 4 Impact of removing outliers on average revenue.
df_no_outliers = df[df['revenue'] <= upper_rev_limit]
avg_revenue_after = df_no_outliers['revenue'].mean()
avg_revenue_after

```

```python
#  Create monthly revenue line chart.
df['order_date'] = pd.to_datetime(df['order_date'])

monthly_revenue = (
    df.groupby(df['order_date'].dt.to_period('M'))['revenue']
      .sum()
      .reset_index()
)

monthly_revenue['order_date'] = monthly_revenue['order_date'].astype(str)

plt.figure()
plt.plot(monthly_revenue['order_date'], monthly_revenue['revenue'], marker='o')
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
# 2 Create top products bar chart.
# 3 Create category revenue pie chart.
# 4 Create heatmap of correlations.
# 5 Build city vs category pivot visualization.
```

```python
# 2 Create top products bar chart.
top_products = (
    df.groupby('product')['revenue']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure()
plt.bar(top_products.index, top_products.values)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
# 3 Create category revenue pie chart.
category_revenue = df.groupby('category')['revenue'].sum()

plt.figure()
plt.pie(category_revenue.values,
        labels=category_revenue.index,
        autopct='%1.1f%%')
plt.title("Revenue Share by Category")
plt.show()
```

```python
# 4 Create heatmap of correlations.
corr = df[['discount_percent', 'unit_price', 'quantity', 'revenue']].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# 5 Build city vs category pivot visualization.
city_category_pivot = pd.pivot_table(
    df,
    values='revenue',
    index='city',
    columns='category',
    aggfunc='sum',
    fill_value=1
)

plt.figure()
plt.imshow(city_category_pivot)
plt.colorbar()
plt.xticks(range(len(city_category_pivot.columns)),
           city_category_pivot.columns,
           rotation=45)
plt.yticks(range(len(city_category_pivot.index)),
           city_category_pivot.index)
plt.title("City vs Category Revenue Heatmap")
plt.tight_layout()
plt.show()
```
